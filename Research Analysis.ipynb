```python
from csv_diff import load_csv, compare
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Demo = pd.read_csv('Files/CR-demographic.csv', header = 1)[9:]
```

Participant's gender distribution

```python
# Histgram of Participant's gender distribution.
x = Demo["Please select your sex:"].value_counts()
y = Demo["Please select your sex:"].value_counts().tolist()
x.plot.barh()
plt.title("Participant's gender distribution")
plt.xlabel("sample")
plt.ylabel("gender") 

for i, v in enumerate(y):
    plt.text(v - 0.8, i - 0.03 , str(v), color='black', fontweight='bold')

plt.savefig(os.path.join('Pics/Research Analysis/sex-hist.png'), dpi=300, format='png', bbox_inches='tight')
plt.show()
```

```python
# Pie chart of Participant's gender distribution.
plt.figure(figsize=[9,7])
Demo["Please select your sex:"].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Gender distribution")
plt.savefig(os.path.join('Pics/Research Analysis/sex-pie.png'), dpi=300, format='png', bbox_inches='tight')
plt.show()
```
