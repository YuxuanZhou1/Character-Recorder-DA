```python
from csv_diff import load_csv, compare
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PA = pd.read_csv('Files/CR-studentexperiences.csv', header = 1)[1:]
PA.head()
```

The distribution about User's special behaviors

```python
fig, ax = plt.subplots()
PA["used others values?"].dropna().astype('int32').plot()
PA["use my terms"].dropna().astype('int32').plot()
PA["refer back to tutorial"].dropna().astype('int32').plot()
ax.legend(['used others values', 'use my terms', 'refer back to tutorial'])

plt.title("Participant's behavior distribution")
plt.xlabel("PID")
plt.ylabel("Times") 

plt.savefig(os.path.join('Pics/Exploring Analysis/Users special behaviors_1.png'), dpi=300, format='png', bbox_inches='tight')
plt.show()
```

```python
df = PA[["used others values?", "use my terms", "refer back to tutorial"]].dropna().astype('int32')
df.plot.bar(rot = 0, stacked = True)

plt.title("Participant's behavior bar chart")
plt.xlabel("PID")
plt.ylabel("Times")

plt.savefig(os.path.join('Pics/Exploring Analysis/Users special behaviors_2.png'), dpi=300, format='png', bbox_inches='tight')
plt.show()
```
